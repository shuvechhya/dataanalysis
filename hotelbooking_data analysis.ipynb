# data visualization

```python
#sns.countplot(x = 'is_canceled',data = df)

plt.title('cancelation status')
plt.bar(['not canceled','canceled '],df['is_canceled'].value_counts())
```

```python
ax = sns.countplot(data = df, x = 'hotel', hue = 'is_canceled')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
resort_hotel=df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel=df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.plot(resort_hotel.index,resort_hotel['adr'],label='Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label='City Hotel')
plt.show()
```

```python
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])
df['month'] = df['reservation_status_date'].dt.month

plt.figure(figsize=(16, 8))
ax = sns.countplot(x='month', hue='is_canceled', data=df)
plt.title('Reservation Status per Month')
plt.xlabel('Month')
```

```python
# avg daily rate for each month 
plt.figure(figsize=(16, 8))
plt.title('adr per month',fontsize=30)
sns.barplot(x='month', y='adr', data=df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index())
plt.show()
```

conclusion:adr per month is highest in jan so is the cancelation from the above figure

```python
#on country basis what is the cancelation rate 
plt.figure(figsize=(16, 8))
plt.title('country reservation cancelation',fontsize=30)
cancelled_data=df[df['is_canceled']==1]
top_10=cancelled_data['country'].value_counts()[:10]
plt.pie(top_10,autopct='%.2f',labels=top_10.index)
plt.show()
```

conclusion: portugal country has the highest cancellation rate 

```python
df['market_segment'].value_counts(normalize=True)
```

```python
cancelled_data['market_segment'].value_counts(normalize=True)
```

```python
cancelled_df_adr=cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date',inplace=True)

not_cancelled_data=df[df['is_canceled']==0]
not_cancelled_df_adr=not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date',inplace=True)

plt.figure(figsize=(20,6))
plt.title('adr')
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'],label='not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'],label='cancelled')
plt.legend()
```

```python
#filter 2016 to 2017 data
cancelled_df_adr=cancelled_df_adr[(cancelled_df_adr['reservation_status_date']>'2016')&(cancelled_df_adr['reservation_status_date']<'2017')]
not_cancelled_df_adr=not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date']>'2016')&(not_cancelled_df_adr['reservation_status_date']<'2017')]
```

```python
plt.figure(figsize=(16,8))
plt.title('adr')
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'],label='not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'],label='cancelled')
plt.legend()
```

filtered data between 2016 and 2017
